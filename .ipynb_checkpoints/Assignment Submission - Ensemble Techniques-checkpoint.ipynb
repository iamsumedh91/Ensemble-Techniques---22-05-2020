    Hence, The Training DataSet has approximately 70% data and The Testing DataSet is approximately 30% of the
    Original Data

```python
print("Original Parkinsons Positives           : {0} ({1:0.2f}%)".format(len(df.loc[df['Status'] == 1]), (len(df.loc[df['Status'] == 1])/len(df.index)) * 100))
print("Original Parkinsons Negatives           : {0} ({1:0.2f}%)".format(len(df.loc[df['Status'] == 0]), (len(df.loc[df['Status'] == 0])/len(df.index)) * 100))
print("")
print("Training Parkinsons Positives           : {0} ({1:0.2f}%)".format(len(Y_train.loc[Y_train[:] == 1]), (len(Y_train.loc[Y_train[:] == 1])/len(Y_train.index)) * 100))
print("Training Parkinsons Negatives           : {0} ({1:0.2f}%)".format(len(Y_train.loc[Y_train[:] == 0]), (len(Y_train.loc[Y_train[:] == 0])/len(Y_train.index)) * 100))
print("")
print("Testing Parkinsons Positives            : {0} ({1:0.2f}%)".format(len(Y_test.loc[Y_test[:] == 1]), (len(Y_test.loc[Y_test[:] == 1])/len(Y_test.index)) * 100))
print("Testing Parkinsons Negatives            : {0} ({1:0.2f}%)".format(len(Y_test.loc[Y_test[:] == 0]), (len(Y_test.loc[Y_test[:] == 0])/len(Y_test.index)) * 100))
```

    Hence, Our Training and Testing Data are almost consistent with the Original Data-Set

# 5. Scaling the Data

```python
# Since the data has many outliers, scaling the Data with Robust Scaler
trans = RobustScaler()
Xs_train = trans.fit_transform(X_train)
Xs_test = trans.fit_transform(X_test)
```

# 6. Training with Standard Classification Algorithms

# 6.a. KNN Classifier:

    Since Scaling affects KNN Classifier, using the scaled data shall provide better model with KNN Classifier

```python
# KNN Classifier: For every data point we take 3 neighbors to classify its value

NNH = KNeighborsClassifier(n_neighbors = 3, weights='distance')
NNH.fit(Xs_train, Y_train)

# Calculating Score for the prediction for KNN

YK_pred = NNH.predict(Xs_test)

print("Accuracy on training set: {:.3f}".format(NNH.score(Xs_train, Y_train)))
print("Accuracy on test set: {:.3f}".format(NNH.score(Xs_test, Y_test)))

# Summarizing the fit of the model
print(); print(metrics.classification_report(Y_test, YK_pred))
```

    Hence, KNN-Classifier gives:
        1. Accuracy          :  88%
        2. Precision Value   :  88%
        3. Recall Value      :  88%

```python
def draw_cm( actual, predicted ):
    cm = metrics.confusion_matrix( actual, predicted, [0,1] )
    sns.heatmap(cm, annot=True,  fmt='.2f', xticklabels = ["Predicted 0", "Predicted 1"] , yticklabels = ["Original 0", "Original 1"] )
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
```

```python
# Confusion Matrix for KNN-Classifier
draw_cm( Y_test, YK_pred )
```

# 6.b. Logistic Regression: 

    Since Scaling does not affect Logistic Regression much, using the original train and test data here

```python
# Logistic Regression: For Logistic Regression, dropping HNR Column, which has negative correlation with 
# Target column, provides better Accuracy, Precision and Recall values
df2 = df.drop(['HNR'], axis=1)
Xlr = df2.drop("Status",axis=1)
Ylr = df2["Status"]
Xlr_train, Xlr_test, Ylr_train,  Ylr_test = train_test_split(Xlr, Ylr,train_size=0.7, test_size=0.3, random_state=10)
```

```python
import warnings
warnings.filterwarnings('ignore')

from sklearn.linear_model import LogisticRegression

# Fitting the model on Train Data
lr_model = LogisticRegression(solver="newton-cg")
lr_model.fit(Xlr_train, Ylr_train)

# Predicting on Test Data
Y_predlr = lr_model.predict(Xlr_test)

# Scoring the Model
lr_modelScore = lr_model.score(Xlr_test, Ylr_test)
print("Accuracy on training set: {:.3f}".format(lr_model.score(Xlr_train, Ylr_train)))
print("Accuracy on test set: {:.3f}".format(lr_model.score(Xlr_test, Ylr_test)))

# Summarizing the fit of the model
print(); print(metrics.classification_report(Ylr_test, Y_predlr))
```

    Hence, Logistic Regression gives:
        1. Accuracy        :  91.5% approx (Better than Training Data)
        2. Precision Value :  92%
        3. Recall Value    :  92%

```python
# Confusion Matrix for Logistic Regression
draw_cm( Y_test, Y_predlr )
```

# 6.c. Support Vector Machines:

        Using the scaled data here, doesn't provide better results than non-scaled data. Hence, using original
        data here

```python
# Support Vector Machines
svc = SVC(C=1000)
svc.fit(X_train, Y_train)

# Predicting on Test Data
Y_predSVM = svc.predict(X_test)


# Scoring the Model
print("Accuracy on training set: {:.3f}".format(svc.score(X_train, Y_train)))
print("Accuracy on test set: {:.3f}".format(svc.score(X_test, Y_test)))

# Summarizing the fit of the model
print(); print(metrics.classification_report(Y_test, Y_predSVM))
```

    Hence, SVM gives:
        1. Accuracy          :  88.1% (Better than Training Data with C=1000)
        2. Precision Value   :  90%

# 7. Stacking Meta Classifier

```python
# Using SKLearn Stacking Classifier for Meta Classifier
base_learners = [
                 ('rf_1', NNH),
                 ('rf_2', lr_model),
                 ('rf_3', svc)
                ]
clf = StackingClassifier(estimators=base_learners, final_estimator=lr_model)
clf.fit(X_train, Y_train)
Y_predStack = clf.predict(X_test)

# Scoring the Model
print("Accuracy on training set: {:.3f}".format(clf.score(X_train, Y_train)))
print("Accuracy on test set: {:.3f}".format(clf.score(X_test, Y_test)))

# Summarizing the fit of the model
print(); print(metrics.classification_report(Y_test, Y_predStack))
```

    Hence, Stacking Classifier gives:
        1. Accuracy           :  90% approx
        2. Precision Value    :  91%
        3. Recall Value       :  90%

# 8. Ensemble Model

```python
# For Prediction with Ensemble Techniques, I am using Microsoft's "LightGBM Classifier Algorithm" which is a
#   Gradient Boosting algorithm which fast and efficient.
lgbm = lgb.LGBMClassifier()

X_train.columns = ["".join (c if c.isalnum() else "_" for c in str(x)) for x in X_train.columns]
X_test.columns = ["".join (c if c.isalnum() else "_" for c in str(x)) for x in X_test.columns]

lgbm.fit(X_train, Y_train)
    
# Making predictions
Y_predlgb = lgbm.predict(X_test)

print("Accuracy on training set: {:.3f}".format(lgbm.score(X_train, Y_train)))
print("Accuracy on test set: {:.3f}".format(lgbm.score(X_test, Y_test)))

# Summarizing the fit of the model
print(); print(metrics.classification_report(Y_test, Y_predlgb))
```

    Hence, LightGBM Boosting Classifier gives:
        1. Accuracy           :  96.6%
        2. Precision Value    :  97%
        3. Recall Value       :  97%

```python
# Confusion Matrix for LightGBM
draw_cm( Y_test, Y_predlgb )
```

```python
# Missclassified Data Points by LightGBM
count_misclassified = (Y_test != Y_predlgb).sum()
print('Misclassified data points for LightGBM: {}'.format(count_misclassified))
```

```python
# Plotting ROC Curve for the Model with LightGBM
ns_probs = [0 for _ in range(len(Y_test))]
lgb_probs = lgbm.predict_proba(X_test)

# Keep probabilities for the positive outcome only
lgb_probs = lgb_probs[:, 1]

# Calculate scores
ns_auc = roc_auc_score(Y_test, ns_probs)
lr_auc = roc_auc_score(Y_test, lgb_probs)

# Summarize scores
print('No Skill: ROC AUC=%.3f' % (ns_auc))
print('LightGBM: ROC AUC=%.3f' % (lr_auc))

# Calculate roc curves
ns_fpr, ns_tpr, _ = roc_curve(Y_test, ns_probs)
lr_fpr, lr_tpr, _ = roc_curve(Y_test, lgb_probs)

# Plotting the roc curve for the model
plt.plot(ns_fpr, ns_tpr, linestyle='--', label='No Skill')
plt.plot(lr_fpr, lr_tpr, marker='.', label='LightGBM')

# Axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')

# show the legend
plt.legend()

# show the plot
plt.show()
```

    Hence, the ROC for LightGBM give a very high value for Area Under the Curve (AUC) i.e. 98.4%
